```python
# Importer la bibliothèque NumPy pour gérer les tableaux et effectuer des calculs numériques
import numpy as np

# Importer pyplot de la bibliothèque matplotlib pour tracer des graphiques
from matplotlib import pyplot as plt

# Importer les fonctions cumulative_trapezoid et interp1d de SciPy
# cumulative_trapezoid : pour effectuer une intégration numérique (règle du trapèze)
# interp1d : pour créer une fonction d'interpolation
from scipy.integrate import cumulative_trapezoid
from scipy.interpolate import interp1d

# Données d'exemple pour le temps et la vitesse (remplacez-les par vos données réelles)
time = np.array([0, 1, 2, 3, 4, 5])  # Temps en secondes
speed = np.array([0, 5, 7, 8, 9, 9])  # Vitesse en m/s

# Tracer la courbe vitesse-temps
plt.plot(time, speed)  # Crée un graphique avec le temps sur l'axe des x et la vitesse sur l'axe des y
plt.title("Courbe vitesse-temps")  # Ajouter un titre au graphique
plt.xlabel("Temps (s)")  # Ajouter un label pour l'axe des x
plt.ylabel("Vitesse (m/s)")  # Ajouter un label pour l'axe des y
plt.grid(True)  # Ajouter une grille pour mieux lire les données
plt.show()  # Afficher le graphique
```

```python
# Calculer la distance cumulative en utilisant la règle du trapèze
# Cela permet de calculer l'intégrale numérique de la vitesse par rapport au temps
distance = cumulative_trapezoid(speed, time, initial=0)
print('la distance cumulative: ', distance)


# Ces fonctions permettront d'estimer les valeurs de temps et de vitesse pour une distance donnée
distance_interp = interp1d(distance, time, fill_value="extrapolate")  # Interpolation distance -> temps
speed_interp = interp1d(distance, speed, fill_value="extrapolate")  # Interpolation distance -> vitesse

# Trouver le temps correspondant à une distance de 31 mètres
time_at_31m = distance_interp(31)

# Trouver la vitesse correspondante à une distance de 31 mètres
speed_at_31m = speed_interp(31)

# Afficher les résultats
print(f"Temps pour atteindre 31 mètres : {time_at_31m:.2f} secondes")
print(f"Vitesse à 31 mètres : {speed_at_31m:.2f} m/s")
```
